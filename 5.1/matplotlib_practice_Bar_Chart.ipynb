```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.bar(x_axis, y_axis, color = "green", label='Boston')
# create labels for x and y axes
plt.xlabel ("Date")
plt.ylabel ("Fare($)")
#create a title
plt.title("PyBer Fare by Month")
#add a legend
plt.legend()
```

```python
#create a horizontal bar chart
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Skill Drill Change the bar colors to magenta.

#Add a legend for the city of Boston.

#Add a title and axes labels.

#Invert the y-axis so that January is at the top.

#create the plot
plt.barh(x_axis, y_axis, color = "magenta", label='Boston')
# create labels for x and y axes
plt.ylabel ("Date")
plt.xlabel ("Fare($)")
#create a title
plt.title("PyBer Fare by Month")
#add a legend
plt.legend()
plt.gca().invert_yaxis()
```

```python
#Create bar charts using object-oriented approach
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a bar plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#create a horizontal bar chart with ax.plt() method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color ="cyan", label="Chicago")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
```
