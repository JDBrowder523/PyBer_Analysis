```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create a plot with ax.plt()
#fig references a figure, ax references axes, ply.subplots() unpacks a tuple function, as.plot plots x and y axes
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create a plot with ax.plt() SECOND METHOD
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create quick graph using object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create a plot with plot show plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot and add a label for a legend
plt.plot(x_axis, y_axis, marker ="*", color = "blue", linewidth=2,label='Boston')
#create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add a legend
plt.legend()
#add a grid
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "D", linewidth = 2, color = "green", label = "Boston")
ax.legend()
ax.grid()
ax.set_title("PyBer Fare By Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0,45)
```
